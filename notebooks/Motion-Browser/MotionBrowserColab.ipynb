## Load metadata

```python
ann_df = pd.read_csv('labels_ds_v2.csv')
print(len(ann_df))
ann_df.head()
```

```python
meta_df = pd.read_csv('c3d_metadata.csv')
print(len(meta_df))
meta_df.head()
```

## search

```python
search_str = 'runs'
results = ann_df[ann_df.text.str.contains(search_str)]
print(len(results))
results
```

## view motion

```python
def display_sample(sample_id):
    meta_rec = meta_df[meta_df.sample_id==sample_id].iloc[0]
    print(meta_rec)
    annotation = ann_df[ann_df.sample_id==sample_id].text.to_list()[0]
    meta_rec = meta_df[meta_df.sample_id==sample_id].iloc[0]
    n_frames = meta_rec.last_frame - meta_rec.first_frame
    print('\n', annotation)
    motion_path = Path(f"/content/motion-viewer/{sample_id:05}_raw.json")
    display_motion(motion_path, n_frames=n_frames)
```

```python
sample_id = 446
display_sample(sample_id)
```
